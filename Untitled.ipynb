```python
from glob import glob
from os.path import expanduser
from sqlite3 import connect
import pathlib
import csv
import emoji
import pandas as pd
from matplotlib import pyplot as plt
from instaloader import ConnectionException, Instaloader, Post
from textblob import TextBlob

path_to_firefox_cookies = "C:\\Users\\Hello\\AppData\\Roaming\\Mozilla\\Firefox\\Profiles\\gtdm9dv9.default-release\\cookies.sqlite"
FIREFOXCOOKIEFILE = glob(expanduser(path_to_firefox_cookies))[0]

loader = Instaloader(max_connection_attempts=1)
loader.context._session.cookies.update(
    connect(FIREFOXCOOKIEFILE).execute(
        "SELECT name, value FROM moz_cookies WHERE host='.instagram.com'"
    )
)

try:
    username = loader.test_login()
    if not username:
        raise ConnectionException()
except ConnectionException:
    raise SystemExit("Cookie import failed. Are you logged in successfully in Firefox?")

loader.context.username = username
loader.save_session_to_file()

instagram = Instaloader(download_pictures=False, download_videos=False,
                        download_video_thumbnails=False, save_metadata=False,
                        max_connection_attempts=0)
instagram.load_session_from_file('majorproject2025')

def scrape_data(url):
    SHORTCODE = str(url[28:39])
    post = Post.from_shortcode(instagram.context, SHORTCODE)
    output_path = pathlib.Path('post_data')
    output_path.mkdir(parents=True, exist_ok=True)
    csv_path = output_path.joinpath(SHORTCODE + '.csv')
    
    with csv_path.open("w", encoding="utf-8", newline="") as post_file:
        field_names = ["post_shortcode", "commenter_username", "comment_text", "comment_likes"]
        post_writer = csv.DictWriter(post_file, fieldnames=field_names)
        post_writer.writeheader()
        
        for x in post.get_comments():
            post_writer.writerow({
                "post_shortcode": post.shortcode,
                "commenter_username": x.owner.username if hasattr(x.owner, 'username') else str(x.owner),
                "comment_text": (emoji.demojize(x.text)).encode('utf-8', errors='ignore').decode() if x.text else "",
                "comment_likes": x.likes_count
            })
    
    print("Done Scraping! CSV saved as:", csv_path)
    return SHORTCODE

sample_url = "https://www.instagram.com/p/DFaLIoXIzVs/"
sample_shortcode = scrape_data(sample_url)
csv_file_path = pathlib.Path('post_data') / (sample_shortcode + '.csv')
df = pd.read_csv(csv_file_path)

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

df['text_polarity'] = df['comment_text'].apply(getPolarity)
df['sentiment'] = pd.cut(df['text_polarity'], [-1, -1e-10, 1e-10, 1], labels=["Negative", "Neutral", "Positive"])

graph1 = df.groupby(['post_shortcode', 'sentiment'], observed=True).size().reset_index(name='count')
graph2 = graph1[graph1['post_shortcode'] == sample_shortcode]

colors = ["#FF0066", "gray", "#00FF00"]
fig, ax = plt.subplots(ncols=1, figsize=(8,6))
ax.bar(graph2["sentiment"], graph2["count"], color=colors)
ax.set_xlabel("Sentiment")
ax.set_ylabel("Number of Comments")
plt.title("Instagram Comment Sentiment for Post " + sample_shortcode)
plt.show()
```

```python
import tkinter as tk
from tkinter import messagebox
import matplotlib.pyplot as plt
import pandas as pd
import pathlib
import csv
import emoji
from instaloader import Instaloader, Post
from textblob import TextBlob

# Load session (Make sure session file is valid)
instagram = Instaloader()
instagram.load_session_from_file('majorproject2025')  # Replace with your session file

# Function to scrape comments and save to CSV
def scrape_data(url):
    shortcode = url.split("/p/")[-1].split("/")[0]  # Extract shortcode
    post = Post.from_shortcode(instagram.context, shortcode)
    
    output_path = pathlib.Path("post_data")
    output_path.mkdir(parents=True, exist_ok=True)
    csv_path = output_path / f"{shortcode}.csv"

    print(f"Saving file to: {csv_path}")  # Debugging print

    with csv_path.open("w", encoding="utf-8", newline="") as post_file:
        field_names = ["post_shortcode", "commenter_username", "comment_text", "comment_likes"]
        post_writer = csv.DictWriter(post_file, fieldnames=field_names)
        post_writer.writeheader()

        comment_count = 0
        for x in post.get_comments():
            comment_text = emoji.demojize(x.text).encode('utf-8', errors='ignore').decode() if x.text else ""
            post_writer.writerow({
                "post_shortcode": post.shortcode,
                "commenter_username": x.owner.username if hasattr(x.owner, 'username') else str(x.owner),
                "comment_text": comment_text,
                "comment_likes": x.likes_count
            })
            comment_count += 1
        
        print(f"Total Comments Scraped: {comment_count}")  # Debugging print

    return shortcode

# Function to calculate polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# Function to analyze sentiment and display results
def analyze_sentiment():
    url = url_entry.get().strip()
    if not url:
        messagebox.showerror("Error", "Please enter a URL")
        return
    
    # Extract shortcode and scrape data
    if "/p/" in url:
        shortcode = url.split("/p/")[-1].split("/")[0]
    else:
        messagebox.showerror("Error", "Invalid Instagram URL format")
        return

    csv_file_path = pathlib.Path("post_data") / f"{shortcode}.csv"
    
    # If file does not exist, scrape data first
    if not csv_file_path.exists():
        print("Data not found, scraping now...")
        shortcode = scrape_data(url)  # Scrape comments
        csv_file_path = pathlib.Path("post_data") / f"{shortcode}.csv"

    # Read CSV file
    if not csv_file_path.exists():
        messagebox.showerror("Error", f"Failed to retrieve data for post: {shortcode}")
        return

    df = pd.read_csv(csv_file_path)

    if 'comment_text' not in df.columns or df.empty:
        messagebox.showerror("Error", "CSV file is missing required data or is empty.")
        return

    # Perform sentiment analysis
    df['text_polarity'] = df['comment_text'].astype(str).apply(getPolarity)
    df['sentiment'] = pd.cut(df['text_polarity'], [-1, -1e-10, 1e-10, 1], labels=["Negative", "Neutral", "Positive"])

    # Count sentiment values
    sentiment_counts = df['sentiment'].value_counts()
    labels = sentiment_counts.index.tolist()
    sizes = sentiment_counts.values.tolist()

    # Display Graphs
    fig, ax = plt.subplots(1, 2, figsize=(12,6))

    colors = ["#FF0066", "gray", "#00FF00"]
    ax[0].bar(labels, sizes, color=colors)
    ax[0].set_xlabel("Sentiment")
    ax[0].set_ylabel("Number of Comments")
    ax[0].set_title(f"Instagram Comment Sentiment ({shortcode})")

    ax[1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
    ax[1].set_title("Sentiment Distribution")

    plt.show()

# GUI Setup
root = tk.Tk()
root.title("Instagram Sentiment Analysis")
root.geometry("400x200")

tk.Label(root, text="Enter Instagram Post URL:").pack(pady=5)
url_entry = tk.Entry(root, width=50)
url_entry.pack(pady=5)

tk.Button(root, text="Analyze", command=analyze_sentiment).pack(pady=10)

root.mainloop()
```
